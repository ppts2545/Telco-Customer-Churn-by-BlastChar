## Setup and Data Loading

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../Data/raws/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## Basic Data Overview

```python
print("Dataset shape:", df.shape)
print("\nColumn data types:")
print(df.dtypes)
print("\nFirst few rows:")
df.head()
```

## Data Type Correction

TotalCharges should be numeric but may contain spaces or invalid values.

```python
print("Data Type of TotalCharges:  ", df["TotalCharges"].dtype)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

print(df["TotalCharges"].unique())

print("\n")

print("Min TotalCharges:", df["TotalCharges"].min())
print("Max TotalCharges:", df["TotalCharges"].max())

print("\n")

print("Skewness of TotalCharges:", df["TotalCharges"].skew())
```

```python
df.describe()
```

```python
print("Null values in each column:\n", df.isnull().sum())
```

```python
print("Duplicate rows in the dataset:", df.duplicated().sum())
```

- ข้อมูลมีความเอียงเล็กน้อย ยังพอไช้ได้

```python
print("Missing values per column:")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
df.select_dtypes(include=['int64', 'float64']).skew()
```

```python
# Ensure plotting columns are numeric in case this cell is run independently
numeric_cols = ["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

sns.kdeplot(df["SeniorCitizen"].dropna(), fill=True)
plt.title("Distribution of Senior Citizens")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

sns.kdeplot(df["tenure"].dropna(), fill=True)
plt.title("Distribution of Tenure")
plt.xlabel("Tenure (months)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

sns.kdeplot(df["MonthlyCharges"].dropna(), fill=True)
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

sns.kdeplot(df["TotalCharges"].dropna(), fill=True)
plt.title("Distribution of Total Charges")
plt.xlabel("Total Charges")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

### Distribution Montly Charge ----
- Monthly charges show a bimodal distribution, indicating two main pricing tiers among customers.

- ลูกค้าจ่ายรายเดือนประมาณช่วงนี้ 18 → 120

- ช่องว่างตรงกลาง 30–45 อาจแปลว่า แพ็กเกจนี้ไม่ popular, หรือไม่มีแพ็กเกจราคาแบบนี้

### Distribution of Tenuree ----
- ลูกค้าใหม่เยอะมาก 0–10 เดือน

- ลูกค้าที่อยู่ นานมากก็เยอะ 65–72 เดือน

- ช่วง 20–50 เดือน ลูกค้าน้อยกว่า

### Distribution of Citizen ----
- ลูกค้าอายุน้อย มี จํานวนมากกว่า กลุ่มลูกค้าผู้สูงวัย

### Distribution of Total Charges ----
ลูกค้าส่วนใหญ่มี Total Charges ต่ำ

Total Charges ต่ำ = จ่ายสะสมน้อย = Tenure น้อย

ข้อมูล Senior Citizen มีความเอียงข้อมูลมาก ต้องปรับวิธีวัด
