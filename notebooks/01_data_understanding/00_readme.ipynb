# 01 Data Understanding - Navigation Hub

> เป้าหมาย (Objective): ทำความเข้าใจโครงสร้างข้อมูลและความเสี่ยงด้านคุณภาพก่อนเข้าสู่ขั้นเตรียมข้อมูล/วิเคราะห์เชิงลึก

> Professional intent: establish dataset understanding and baseline data quality before preparation and modeling.

---

## Module Structure

### [01_data_overview.ipynb](01_data_overview.ipynb)
**ภาพรวมข้อมูลและบริบทธุรกิจ | Data overview and business context**
- Dataset shape and schema
- Target (Churn) distribution
- Column definitions and meaning

### [02_data_quality.ipynb](02_data_quality.ipynb)
**ตรวจสอบคุณภาพข้อมูล | Data quality profiling**
- Data type checks and conversion diagnostics
- Missing values and duplicates
- Categorical value validation

---

## Recommended Workflow

1. Run **01_data_overview.ipynb**
2. Run **02_data_quality.ipynb**
3. Continue to **Data Preparation**: [../02_data_preparation/01_business_rules_validation.ipynb](../02_data_preparation/01_business_rules_validation.ipynb)

---

## Why This Structure

- **Understanding** = รู้จักข้อมูลและความเสี่ยง (know the data and risks)
- **Preparation** = ใช้ business rules และสร้าง model-ready features
- **EDA** = วิเคราะห์พฤติกรรมและความสัมพันธ์ของตัวแปร
- **Modeling/Evaluation** = เทรนโมเดลและประเมินผลเชิงเทคนิค + เชิงธุรกิจ

โครงสร้างนี้สอดคล้องกับ workflow ในทีม Data/Analytics จริง และเหมาะสำหรับ portfolio สมัครงาน.
