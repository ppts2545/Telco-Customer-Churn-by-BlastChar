# 01 Data Quality Check

This notebook handles:
- Data loading
- Data type validation and conversion
- Missing value detection
- Data export for downstream analysis

## Setup and Data Loading

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../Data/raws/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## Basic Data Overview

```python
print("Dataset shape:", df.shape)
print("\nColumn data types:")
print(df.dtypes)
print("\nFirst few rows:")
df.head()
```

## Data Type Correction

TotalCharges should be numeric but may contain spaces or invalid values.

```python
print("Data Type of TotalCharges:  ", df["TotalCharges"].dtype)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

print(df["TotalCharges"].unique())

print("\n")

print("Min TotalCharges:", df["TotalCharges"].min())
print("Max TotalCharges:", df["TotalCharges"].max())
```

## Missing Value Check

```python
print("Missing values per column:")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())
```

## Duplicate Check

```python
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
df["ChargeGroup"] = pd.qcut(df["MonthlyCharges"], 4)
```

```python
# Save cleaned data for next step
df.to_csv('../../data/clean/cleaned_telco_churn.csv', index=False)
print("✓ Cleaned data saved to: data/clean/cleaned_telco_churn.csv")
```
