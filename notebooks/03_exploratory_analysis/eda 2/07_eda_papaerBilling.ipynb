```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import ttest_ind

df = pd.read_csv('../../../data/processed/telco_customer_churn_clean.csv')
```

```python
pd.crosstab(df["PaperlessBilling"], df["Churn"], normalize="index")
```

```python
sns.countplot(data=df, x="PaperlessBilling", hue="Churn")
plt.title("Churn by Paperless Billing")
plt.xlabel("Paperless Billing")
plt.ylabel("Count")
plt.legend(title="Churn", loc="upper right")
plt.show()
```

```python
ct = pd.crosstab(df["PaperlessBilling"], df["InternetService"])

ct.plot(kind="bar", stacked=True)
plt.ylabel("Count")
plt.title("Internet Service by Paperless Billing")
plt.show()
```

```python
no_internet = df[df["InternetService"] == "No"]

no_internet["PaperlessBilling"].value_counts().plot(kind="bar")

plt.ylabel("Count")
plt.title("Customers without Internet Service")
plt.show()
```

```python
sns.boxplot(x="PaperlessBilling", y="InternetService", data=df)
plt.title("Internet Service by Paperless Billing")
plt.xlabel("Paperless Billing")
plt.ylabel("Internet Service")
plt.show()

print("Mean Monthly Charges by Paperless Billing:")
df.groupby("PaperlessBilling")[""].mean()
```

```python
print("Proportion of No Internet Service by Paperless Billing:")
print(df[df["PaperlessBilling"] == "No"]["InternetService"].eq("No").mean())
```

### Insight

    Customers who do not use paperless billing tend to have no internet service.
    As a result, their monthly charges are lower, which is associated with a lower churn rate.
