# 01 Hypothesis Testing

Template for key statistical tests in churn analysis.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind

df = pd.read_csv('../../data/processed/telco_customer_churn_clean.csv')
```

## Test 1: Categorical vs Churn (Chi-square)
Example: Contract vs Churn

```python
table = pd.crosstab(df['Contract'], df['Churn'])
chi2, p, dof, expected = chi2_contingency(table)
print(f'chi2={chi2:.4f}, p-value={p:.6f}, dof={dof}')
```

## Test 2: Numeric vs Churn (t-test)
Example: MonthlyCharges by churn group

```python
y = df['Churn'].map({'No': 0, 'Yes': 1}) if df['Churn'].dtype == 'object' else df['Churn']
group_no = df.loc[y == 0, 'MonthlyCharges'].dropna()
group_yes = df.loc[y == 1, 'MonthlyCharges'].dropna()
t_stat, p_val = ttest_ind(group_no, group_yes, equal_var=False)
print(f't-stat={t_stat:.4f}, p-value={p_val:.6f}')
```
