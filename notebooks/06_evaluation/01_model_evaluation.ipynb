## 3) Core Classification Metrics

```python
def evaluate_classification(y_test, y_pred, y_prob=None):
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
    }
    if y_prob is not None:
        metrics['roc_auc'] = roc_auc_score(y_test, y_prob)
    return pd.Series(metrics)

# Example
# metrics = evaluate_classification(y_test, y_pred, y_prob)
# display(metrics.to_frame('value'))
```

## 6) Business Impact Estimation

ตัวอย่างกรอบคิด: หาก retention campaign targeting top-risk customers แล้วรักษาได้ X%
ให้ประเมิน revenue saved และ churn reduction uplift

```python
def estimate_business_impact(n_targeted, save_rate, avg_monthly_revenue):
    retained_customers = n_targeted * save_rate
    monthly_revenue_saved = retained_customers * avg_monthly_revenue
    return {
        'retained_customers': retained_customers,
        'monthly_revenue_saved': monthly_revenue_saved
    }

# Example
# impact = estimate_business_impact(n_targeted=500, save_rate=0.2, avg_monthly_revenue=70)
# impact
```
