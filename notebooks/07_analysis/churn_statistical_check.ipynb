## Check VALIDEATE Feature insight from model

(Validate Insigth from Model) before we summurize data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../data/processed/telco_customer_churn_clean.csv')
```

```python
df["Churn"] = df["Churn"].map({"No":0, "Yes":1})
```

#### Anlyst Contract Feature 

```python
df.groupby('Contract')['Churn'].mean()
```

### Check Feature ที่ โมเดล คาดการณ์ มีผล ต่อ Churn 

```python
from scipy.stats import chi2_contingency

features = [
    "InternetService",
    "MultipleLines",
    "PaperlessBilling",
    "StreamingMovies",
    "StreamingTV",
    "tenure_group"
]

for col in features:

    table = pd.crosstab(df[col], df["Churn"])
    chi2, p, dof, exp = chi2_contingency(table)

    print(f"{col}  p-value = {p}")
```
